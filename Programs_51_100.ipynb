```python
#59.Count the number of vowels and consonants in a string.
def count_vowels_and_consonants(s):
    vowels = "aeiouAEIOU"
    vowel_count = 0
    consonant_count = 0

    for char in s:
        if char.isalpha():
            if char in vowels:
                vowel_count += 1
            else:
                consonant_count += 1

    return vowel_count, consonant_count

input_string = "welcome to python programming!"
vowels, consonants = count_vowels_and_consonants(input_string)

print(f"Number of vowels: {vowels}")
print(f"Number of consonants: {consonants}")
```

```python
#64.Merge two sorted lists into one sorted list.

def merge_sorted_lists(list1, list2):
    merged_list = []
    i, j = 0, 0

    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1

    # If there are remaining elements in either list, append them
    merged_list.extend(list1[i:])
    merged_list.extend(list2[j:])

    return merged_list
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]
merged_result = merge_sorted_lists(list1, list2)

print("List 1:", list1)
print("List 2:", list2)
print("Merged and sorted list:", merged_result)
```

```python
#65.Check if a list is sorted in ascending order.
def is_sorted_ascending(lst):
    n = len(lst)
    for i in range(n-1):
        if lst[i] > lst[i+1]:
            return False
    return True
sorted_list = [1, 2, 3, 4, 5]
unsorted_list = [3, 1, 4, 1, 5, 9, 2, 6]

print("Is the sorted list ascending?", is_sorted_ascending(sorted_list))
print("Is the unsorted list ascending?", is_sorted_ascending(unsorted_list))
```

```python
#88.Merge two dictionaries and handle duplicate keys by concatenating values into lists.
def merge_and_concat(dict1, dict2):
    merged_dict = {}
    for key, value in dict1.items():
        merged_dict[key] = [value]

    for key, value in dict2.items():
        if key in merged_dict:
            merged_dict[key].append(value)
        else:
            merged_dict[key] = [value]

    return merged_dict
dict_a = {'a': 1, 'b': 2, 'c': 3}
dict_b = {'b': 4, 'c': 5, 'd': 6}

result = merge_and_concat(dict_a, dict_b)

print("Merged Dictionary with Concatenated Values:")
print(result)
```

```python
#94.Find the first non-repeated character in a string.
def first_non_repeated_char(input_string):
    char_count = {}


    for char in input_string:
        char_count[char] = char_count.get(char, 0) + 1

    for char in input_string:
        if char_count[char] == 1:
            return char

    # Return None if there is no non-repeated character
    return None


my_string = "programming"

result = first_non_repeated_char(my_string)

if result:
    print(f"The first non-repeated character is: {result}")
else:
    print("No non-repeated character found.")
```

```python
#97.Remove all punctuation from a string.
import string

def remove_punctuation(input_string):

    translation_table = str.maketrans("", "", string.punctuation)

    result = input_string.translate(translation_table)

    return result

original_string = "Hello, World! This is an example string with punctuation."
result = remove_punctuation(original_string)

print("Original String:", repr(original_string))
print("String without Punctuation:", repr(result))
```

```python
#100.Calculate the average of values in a dictionary.
def calculate_average(dictionary):
    values = dictionary.values()

    # Check if the dictionary is not empty to avoid division by zero
    if values:
        return sum(values) / len(values)
    else:
        return 0  # Return 0 if the dictionary is empty

my_dict = {'a': 10, 'b': 20, 'c': 30}

average = calculate_average(my_dict)

print("Dictionary:", my_dict)
print("Average of Values:", average)





```
